```python
import random

ans = []
for j in range(3):
    ans.append(random.randint(0, 9))

while True:
    hit = 0
    ball = 0
    prediction = []

    # --- 1. まず3つ全部入力してもらう ---
    for i in range(3):
        # inputはint()で数値に変えておくと、ansの数値と比較しやすくなります
        user_guess = int(input(f"{i+1}桁目の数字を入力(0~9) >> "))
        prediction.append(user_guess)

    # --- 2. 3つ揃ったところで、一気に判定する ---
    for i in range(3):
        if prediction[i] in ans:
            if prediction[i] == ans[i]:
                hit += 1
            else:
                ball += 1

    # --- 3. 数えた結果を表示する ---
    print(f"{hit}ヒット、{ball}ボール") # ここをhitとballに変えるのがコツ！

    if hit == 3: 
        print("クリア！")
        break
    else:
        continue_input = input("ゲームを続けますか？ (y/n) >> ") 
        if continue_input == "n":
            print(f"ゲームを終了します。正解は {ans} でした！")
            break
```
